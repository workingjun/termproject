```python
import requests
from bs4 import BeautifulSoup
from coupang_selenium import SearchManager as SM
from selenium import webdriver
from selenium.webdriver.common.by import By
import chromedriver_autoinstaller
import subprocess
from coupang_selenium import Setting_Options

infro='C:\Program Files\Google\Chrome\Application\chrome.exe --remote-debugging-port=9222 --user-data-dir="C:\chrometemp"'
subprocess.Popen(rf'{infro}')
keyword = "지갑"
options = webdriver.ChromeOptions()
options.add_experimental_option("debuggerAddress", "127.0.0.1:9222")

chromedriver_autoinstaller.install()
driver = webdriver.Chrome(options=options)
driver.implicitly_wait(3)
Setting_options=Setting_Options(driver)
Setting_options.getting_keyword(keyword)
######################################## 쿠팡 검색하기 ######################################################
## 페이지 수 
page_num = 1

## 키워드 입력 


## 페이지 수만큼 반복 
while page_num<2:
 
        raw = Setting_options.page_source()

        ## html을 입력으로 넣고 파싱함 -> 구조 구문 분석 
        search = BeautifulSoup(raw, 'html.parser')
        ########################################################## 여기까진 공식 

        ## 태그 네임이 ul이고 id가 productList인 요소 찾기
        box = search.find('ul', {'id' : 'productList'})
        ## 태그 네임이 li인 모든 요소들 찾기 -> 결과는 변수에 리스트로 저장됨
        all_product = box.find_all('li', {'class': 'search-product'})

        print('페이지:', page_num, "<쿠팡" + keyword + "결과>:")

        
        for i, product in enumerate(all_product):

                ## 이미지 링크 찾기 
                imagefind = product.find("img", {'class' : 'search-product-wrap-img'})

                ## 예외처리
                if imagefind.get("data-img-src") is None:
                        imageLink = "http:" + imagefind.get('src')
                else:
                       imageLink = "http:" + imagefind.get("data-img-src")

                ## 제목
                title = product.find('div', {'class' : 'name'})

                ## 가격
                price = product.find('strong', {'class' : 'price-value'})

                ## 별점
                point = product.find('span', {'class' : 'rating-total-count'})
                
                ## 별점 예외처리
                if point is None:
                       point=''

                ## 링크 가공
                link2 = product.find('a')['href']
                link = "https://www.coupang.com" + link2
                
                print(i, '.')
                print("이미지 링크:", imageLink)
                print('이름 : ', title.text.strip())
                print('가격 : ', price.text)
                
                if isinstance(point, str):
                        print('별점 : ', point)
                else:
                       print('별점 : ', point.text)

                print('링크:', link)
                print()
                      
        print()
        page_num+=1
```
